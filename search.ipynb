```python
from icrawler.builtin import GoogleImageCrawler
import os
import re

characters = ["A-Train", "Kimiko", "Frenchie", "Soldier Boy"]

for character in characters:
    # Формируем точный поисковый запрос
    query = f"{character} character from The Boys"

    # Убираем пробелы и спец. символы из названия папки
    folder_name = re.sub(r'[^a-zA-Z0-9]', '_', character)

    # Убедимся, что папка существует
    os.makedirs(f'images/{folder_name}', exist_ok=True)

    # Запускаем краулер
    crawler = GoogleImageCrawler(storage={'root_dir': f'images/{folder_name}'})
    crawler.crawl(
        keyword=query,
        max_num=100,
        file_idx_offset=0,
        overwrite=False  # не перезаписывает старые изображения
    )

    print(f"[+] Загрузка завершена для {character}")
```
